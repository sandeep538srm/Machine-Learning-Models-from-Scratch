# 1. Types of distances

```python
def eucledian(p1,p2):
    return (((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2)+((p1[2]-p2[2])**2))**.5

def manhatten(p1,p2):
    return (abs(p1[0]-p2[0])+abs(p1[1]-p2[1])+abs(p1[2]-p2[2]))

def minkowski(p1,p2,k):
    return (((p1[0]-p2[0])**k)+((p1[1]-p2[1])**k)+((p1[2]-p2[2])**k))**(1/k)

if __name__=="__main__":
    print("Enter the coordinates of first point which has 3 dimensions")
    p1=list(map(int,input().split()))
    print("Enter the coordinates of second point which has 3 dimensions")
    p2=list(map(int,input().split()))
    k=int(input("Enter the value of K in Minkowski Distance:"))
    print(f'First point is: ({p1[0]},{p1[1]},{p1[2]})')
    print(f'Second point is: ({p2[0]},{p2[1]},{p2[2]})')
    print(f'Eucledian Distance between ({p1[0]},{p1[1]},{p1[2]}), ({p2[0]},{p2[1]},{p2[2]}) is: {eucledian(p1,p2)}')
    print(f'Manhatten Distance between ({p1[0]},{p1[1]},{p1[2]}), ({p2[0]},{p2[1]},{p2[2]}) is: {manhatten(p1,p2)}')
    print(f'Minkowski Distance between ({p1[0]},{p1[1]},{p1[2]}), ({p2[0]},{p2[1]},{p2[2]}) is: {minkowski(p1,p2,k)}')
```

# Distances between points in Iris Dataset

```python
from sklearn import datasets
import pandas as pd

x=(datasets.load_iris())
print(x)
```

```python
y=pd.DataFrame(x.data)
length=len(y)
whole_data=pd.concat([pd.DataFrame(x.data),pd.DataFrame(x.target)],axis=1)
print(whole_data)
array=[[0 for i in range(length)] for i in range(length)]
for i in range(length):
    for j in range(length):
        array[i][j]=(((y[0][i]-y[0][j])**2)+((y[1][i]-y[1][j])**2)+((y[2][i]-y[2][j])**2)+((y[3][i]-y[3][j])**2))**.5
print(array)
```

 If we use regression line, we will get the following

```python
import numpy as np
ab=sample_data.iloc[:100,0]
cd=sample_data.iloc[:100,1]
b0,b1=0,0
n=100
learning_rate=0.0001
print(ab)
print(cd)
for i in range(4000):
    y_computed=b0+b1*ab
    for j in range(np.size(cd)):
        b0=b0+(learning_rate)*(cd[j]- y_computed[j])
        b1=b1+learning_rate*(cd[j]- y_computed[j])*ab[j]
    error=(np.sum((cd-y_computed)**2))/n
    if(error<1e-6):
        break
print(i,b0,b1)
```
